## Preprocessing

```python
# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Additional libraries
from pydrive2.auth import GoogleAuth
from pydrive2.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
```

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import pandas and read the charity_data.csv from the provided cloud URL.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# ID the following in the dataset: what variable(s) are the target(s) for your model? what variable(s) are the feature(s) for your model?
# target: IS_SUCCESSFUL
# features: APPLICATION_TYPE, AFFILIATION, CLASSIFICATION, USE_CASE, ORGANIZATION, STATUS, INCOME_AMT, SPECIAL_CONSIDERATIONS, ASK_AMT
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
application_df
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
app_type_counts = application_df['APPLICATION_TYPE'].value_counts()
app_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

# cutoff value in sample output: 528
application_types_to_replace = list(app_type_counts.iloc[8:].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_values = application_df['CLASSIFICATION'].value_counts()
classification_values
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_values_over1 = classification_values.loc[classification_values > 1]
classification_values_over1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`

# cutoff value in sample output: 1883
classifications_to_replace = list(classification_values.iloc[5:].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies = pd.get_dummies(application_df)
dummies
```

```python
# Split our preprocessed data into our features and target arrays
# target: IS_SUCCESSFUL
# features: everything else
y = dummies['IS_SUCCESSFUL'].values
X = dummies.drop(columns = 'IS_SUCCESSFUL').values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Step 3: Optimize the Model

original model accuracy = 72.63%

*attempt #1*

```python
# keep layer 1 nodes = 8, adjust layer 2 nodes = 4, change output layer activation to sigmoid
input_features = len(X_train[0])
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 4

nn1 = tf.keras.models.Sequential()

# First hidden layer
nn1.add(tf.keras.layers.Dense(units = hidden_nodes_layer1, input_dim = input_features, activation = "relu"))

# Second hidden layer
nn1.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, input_dim = input_features, activation = "relu"))

# Output layer
nn1.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid"))

# Check the structure of the model
nn1.summary()
```

```python
nn1.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
train_model1 = nn1.fit(X_train_scaled, y_train, epochs = 30)
```

```python
# evaluate model 1
model_loss, model_accuracy = nn1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

attempt #1 accuracy: 72.45%

*attempt #2*

```python
# adjust layer 1 nodes = 10, adjust layer 2 nodes = 5, change hidden layers' activation to tanh
input_features = len(X_train[0])
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 4

nn2 = tf.keras.models.Sequential()

# First hidden layer
nn2.add(tf.keras.layers.Dense(units = hidden_nodes_layer1, input_dim = input_features, activation = "tanh"))

# Second hidden layer
nn2.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, input_dim = input_features, activation = "tanh"))

# Output layer
nn2.add(tf.keras.layers.Dense(units = 1, activation = "relu"))

# Check the structure of the model
nn2.summary()
```

```python
nn2.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
train_model2 = nn2.fit(X_train_scaled, y_train, epochs = 30)
```

```python
# evaluate model 2
model_loss, model_accuracy = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

attempt #2 accuracy: 72.12%

*attempt #3*

```python
# increase layer 1 nodes = 10, adjust layer 2 nodes = 6, change hidden layer activation to relu/tanh
# change output layer back to relu
input_features = len(X_train[0])
hidden_nodes_layer1 = 10
hidden_nodes_layer2 = 6

nn3 = tf.keras.models.Sequential()

# First hidden layer
nn3.add(tf.keras.layers.Dense(units = hidden_nodes_layer1, input_dim = input_features, activation = "relu"))

# Second hidden layer
nn3.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, input_dim = input_features, activation = "tanh"))

# Output layer
nn3.add(tf.keras.layers.Dense(units = 1, activation = "relu"))

# Check the structure of the model
nn3.summary()
```

```python
nn3.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
train_model3 = nn3.fit(X_train_scaled, y_train, epochs = 30)
```

```python
model_loss, model_accuracy = nn3.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

attempt #3 accuracy: 72.33%
